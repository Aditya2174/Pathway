```python
import pandas as pd
```

```python
df = pd.read_json('CUAD_v1/CUAD_v1.json')
```

```python
print(df.shape)
df.head()
```

```python
df_data = pd.DataFrame(df['data'].values.tolist())
print(df_data.shape)
df_data.head()
```

```python
df_data.columns
```

```python
idx = 0
print(df_data['title'][idx])
print()
for e in df_data['paragraphs'][idx][0]['qas']:
    # Print in green color
    print('\033[92m' + 'Question:' + '\033[0m')
    print(e['question'])
    # Print in blue color
    print('\033[94m' + 'Answer:' + '\033[0m')
    print('\n'.join([temp['text'] for temp in e['answers']]))
    print('\n' + '='*50 + '\n')
```
